```python
mean_by_person = df.groupby('id person')['Signal'].mean()
print(mean_by_person)
```

```python
mean_by_person = df.groupby('Labels')['Signal'].mean()
print(mean_by_person)
```

```python
summary = df.groupby(['Labels','id person']).agg(
    mean_signal = ('Signal','mean'),
    std_signal = ('Signal','std'),
    count = ('Signal','count')
)
print(summary)
```

```python
print(df.dtypes)
```

```python
summary.to_csv('summary_statistics.csv', index=False)
```
